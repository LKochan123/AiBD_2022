#### Lab5 Tier Protocol i Tidy Data

```python
import pandas as pd
import re

N = len('Kochańczyk')
print(f'Do analizy danych mam zbiór nr: {N % 5 + 1}')
```

##### Wczytanie pliku tekstowego oraz rozdzielenie danych na odpowiednie kolumny:

```python
# Wczytanie pliku tekstowego
path = '../Original Data/weather.txt'
with open(path) as f:
    data_from_file = f.readlines()

regex = re.compile('(\s+I|S|OI|OS|\s+SI)')
data = []

# Obróbka danych
for record in data_from_file:
    record = regex.sub(" ", record)
    data_in_first_item = ["MX17004", record[11:15], record[15:17], record[17:21]]
    days = record[21:]
    days = ['---' if temperature == '9999' else float(temperature) for temperature in re.findall('[0-9]+', days)]
    data.append(data_in_first_item + days)

# Wczytanie danych do pandas oraz zadeklarowanie nazw kolumn
df = pd.DataFrame(data)
df.columns = ["id", "year", "month", "type"] + [f'd{i}' for i in range(1, 32)]
df
```

##### Wyświetlenie danych z pierwszych 5 miesięcy 2010 roku:

```python
# Pomocnicza funkcja która wybiera odpowiednie rekodry z danego roku z pierwszymi N miesiącami
def select_year_till_month(year: str, last_month: int, df: pd.DataFrame):
    for index, row in df.iterrows():
        if row['year'] != year or float(row['month']) > last_month or row['type'] == 'PRCP':
            df.drop(index, axis=0, inplace=True)
    return df

df = select_year_till_month('2010', 5, df)
df
```

##### Melt:

```python
# Przygotowanie danych do nowego DataFrame
def melt_data(df: pd.DataFrame):
    melted_rows = []
    for _, row in df.iterrows():
        id, year, month, element = row['id'], row['year'], row['month'], row['type']
        for d in range(1, 32):
            if row[f'd{d}'] != '---':
                day = f'0{d}' if d < 10 else f'{d}'
                date = f'{year}-{month}-{day}'
                value = int(row[f'd{d}']) / 10
                melted_rows.append([id, date, element, value])

    # Sortowanie po dacie, żeby tmax tmin było na zmiane, ułatwi to ostatni krok
    melted_rows.sort(key=lambda x: x[1])
    melted_df = pd.DataFrame(melted_rows, columns=['id', 'date', 'element', 'value'])
    return melted_df

df = melt_data(df)
df
```

#### Tidy, ostateczne uporządkowanie danych:

```python
def tidy_data(df: pd.DataFrame):
    tided_rows = []
    N = len(df.index)

    for i in range(0, N, 2):
        id = df['id'][i]
        date = df['date'][i]
        tmax = df['value'][i]
        tmin = df['value'][i+1]

        tided_rows.append([id, date, tmax, tmin])
        tided_df = pd.DataFrame(tided_rows, columns=['id', 'date', 'tmax', 'tmin'])

    return tided_df

df = tidy_data(df)
df
```

#### Zapisanie danych do folderu Analysis Data

```python
path = '../Analysis Data/weather_result.csv'
df.to_csv(path, index=False)
```
