### Zadanie 3

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 5

N = 100

x1 = np.linspace(-1, 1, N)
x2 = np.linspace(-6, 6, N)
x3 = np.linspace(0, 5, N)

def draw_plot(x):
    y = f(x)
    plt.plot(x, y,label='x**2+5')
    plt.title('Wykres funkcji f(x)')
    plt.legend()
    plt.grid()
    plt.show()
    
draw_plot(x1)
draw_plot(x2)
draw_plot(x3)
```
